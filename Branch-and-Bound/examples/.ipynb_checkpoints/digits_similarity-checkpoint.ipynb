# DTW

```python
dist = 'eucl'

if dist == 'eucl':
    # Euclidean
    d = lambda a, b: np.sum([(a-b)**2 for x in (a-b)])
    f_d = lambda DD, a, b: np.sqrt(DD**2 + d(a, b))
    final_operator = lambda x: np.sqrt(x)

elif dist == 'manh':
    # Manhattan
    d = lambda a, b : np.abs(a-b)
    f_d = lambda DD, a, b: DD + d(a,b)
    final_operator = lambda x: x

else:
    # Minkowksi
    d = lambda a, b : np.abs(a-b)
    f_d = lambda DD, a, b: max(DD, d(a,b))
    final_operator = lambda x: x
```

```python
how_many = 100
X_trial = []
y_trial = []
for c in np.unique(outputs_MNIST):
    inner_list_d = []
    inner_list_l = []
    for d, l in zip(inputs_MNIST, outputs_MNIST):
        if (len(inner_list_d)<how_many) and (l==c):
            inner_list_d.append(d)
            inner_list_l.append(l)
    X_trial += inner_list_d
    y_trial += inner_list_l
```

```python
from scipy.spatial.distance import euclidean
from fastdtw import fastdtw
```

```python
a4_dims = (17, 10)
fig, axes = plt.subplots(2, 5, figsize=a4_dims)
i = 0

for digit in tqdm(range(10)):
    
    # select a signature corresponding to selected digit
    index = [i for i,l in zip(range(len(y_trial)), y_trial) if l==digit][np.random.randint(how_many)]
    element = X_trial[index]

    # calculate distance between selected path signature and all the others in the dataset
#     distances = [Warp(x, element, d, f_d, final_operator, plot_2d=True).cost for x in X_trial]
    distances = [fastdtw(x, element, dist=euclidean)[0] for x in X_trial]
    
    data_digit = [d for d,l in zip(distances, y_trial) if l==digit]
    data_allOthers = [d for d,l in zip(distances, y_trial) if l!=digit]
    
    if i<5:
        ii = 0
    else:
        ii = 1
    axes[ii, digit % 5].hist(data_digit, 100, density=True, histtype='bar', color='blue', label='same digit')
    axes[ii, digit % 5].hist(data_allOthers, 100, density=True, histtype='bar', color='red', label='all other digits', alpha=0.6)
    axes[ii, digit % 5].legend(prop={'size': 10})
    axes[ii, digit % 5].set_title('Digit {}'.format(digit))
    i += 1

plt.suptitle('Distribution of distances')
plt.show()
```

# Clustering: Warped rough p-variation distance

```python
p_var = 1.5
sig_depth = 2
```

### Pre-computing distance matrix

```python
rough_warping_pairwise_dist = {}
rough_warp_problem = {}
rough_warp_solution = {}

for i_p, p in tqdm(enumerate(flatten_paths), total=len(flatten_paths)):
    
    inner_rough_warping_dict = {}
    inner_prob = {}
    inner_soln = {}
    
    for i_q, q in enumerate(flatten_paths[i_p:]):
        
        # define problem
        problem = BnBWarping(x=p, y=q, p=p_var, depth=sig_depth, norm='l1', root_node=(0,0), bc=4, 
                             plot_2d=True, pvar_dist_mem=None, cache_size=1024)
        
        # solution via BnB algo
        results = pybnb.Solver().solve(problem, log=None, queue_strategy='depth')
        
        # store results
        inner_rough_warping_dict[i_p+i_q] = results.objective
        inner_prob[i_p+i_q] = problem
        inner_soln[i_p+i_q] = results
        
    rough_warping_pairwise_dist[i_p] = inner_rough_warping_dict
    rough_warp_problem[i_p] = inner_prob
    rough_warp_solution[i_p] = inner_soln
```

```python
i = 0
j = 1

problem = rough_warp_problem[i][j]
results = rough_warp_solution[i][j]

best_warp = results.best_node.state
_, optimal_partition = problem.distance(results.best_node.state, optim_partition=True)
problem.plot_alignment([best_warp[k] for k in optimal_partition])
plt.show()
```

```python
df_rough_warping = pd.DataFrame.from_dict(rough_warping_pairwise_dist).fillna(0.) \
                   + pd.DataFrame.from_dict(rough_warping_pairwise_dist).fillna(0.).T
```

```python
df_rough_warping.to_csv('../../data/rough_warping_pairwise_distances_p_{}.csv'.format(p_var))
```
